**K-means Algorithm**:

1. Initialization: Randomly sample $k$ colors from the input image. These are the initial $k$ means $\mu_1,\mu_2,...,\mu_k$.


2. For each pixel in the image, assign it to its nearest mean given by $$c^{(i)} := \textrm{argmin}_{j}\left \|x^{(i)} - \mu_j \right \|^2$$


3. Update the means using the pixel assignments from Step 2. $$\mu_j := \frac{\sum_{i}^{n}1\left \{c^{(i)} = j  \right \}x^{(i)}}{\sum_{i}^{n}1\left \{c^{(i)} = j  \right \}}$$


4. Repeat Steps 2 and 3 until convergence.

```python
from sklearn.cluster import MiniBatchKMeans#since we are dealing with high res images
```

```python
kmeans=MiniBatchKMeans(16).fit(img_data)
#newcolors=kcolors Now we recolor each one of the points in original image with latest 
#cluster assignments- colors we get from kmeans
k_colors=kmeans.cluster_centers_[kmeans.predict(img_data)]
y=plot_utils(img_data,colors=k_colors,title="Reduced color space: 16 colors")
y.colorSpace()
```

### Task 5: K-means Image Compression with Interactive Controls
---

```python
img_dir='images/'
```

```python
@interact
def color_compression(image=os.listdir(img_dir),k=IntSlider(min=1,max=256,step=1,value=16,
                                                           continuous_update=False,
                                                           layout=dict(width='100%'))):
    #os.dir to get a list of images i a particular directory, 
    #Ipywidget IntSlider k=clusters
    input_img=io.imread(img_dir+image)#reads image as numpy array
    img_data=(input_img/255.0).reshape(-1,3)#reshape into 3 long vectors one each for rgb
    
    kmeans=MiniBatchKMeans(k).fit(img_data)
    k_colors=kmeans.cluster_centers_[kmeans.predict(img_data)]
    #from returned kmeans object we look at the cluster centers for which we use k means 
    #to predict new color assignments on original image data
    #we load larger image into memory and replace each pixel with nearest of centroid 
    #colors we found in the kmeans clustering algorithm
    
    k_img=np.reshape(k_colors,(input_img.shape))
    
    fig,(ax1,ax2)=plt.subplots(1,2)#1 row, 2 columns
    fig.suptitle('K-Means Image Compression', fontsize=20)
    
    ax1.set_title('Compressed Image')
    ax1.set_xticks([])
    ax1.set_yticks([])
    ax1.imshow(k_img)
    
    ax2.set_title('Original (16,777,216 colors)')#256^3
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax2.imshow(input_img)   
    
    plt.subplots_adjust(top=0.85)#adjust heading from the top so that the headings do not overlap
    plt.show()
```
